```python
import pandas as pd
from tqdm import trange
```

```python
df=pd.read_csv("panel_data_1021/after_merge_1021.csv")
# df 是你的面板数据
# 1. 筛选
sub = df[df['pro_code'] == 31] 

# 2. 按 hs_residence 计数
count = sub['hs_residence'].value_counts()

# 如果想直接得到 DataFrame，可再 reset_index
count_df = count.reset_index(name='count')
```

```python
count_df["code"]=None
for i in range(len(count_df)):
    count_df.loc[i,"code"]=str(int(count_df.loc[i,"hs_residence"]))+"0000"
count_df
```

```python
count_df.to_excel("DataScience/migrate_to_Shanghai.xlsx",index=None)
```
